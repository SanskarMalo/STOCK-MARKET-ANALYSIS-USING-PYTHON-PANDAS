```python
import pandas as pd
df=pd.read_csv('df10.csv',parse_dates=['Date'],index_col=["Date"])
df.head(50)
```

```python
df.index
```

```python
df.loc['2022-07-03':'2022-08-03'] # shows the Daily stock price between July and August 3rd
```

```python
df.loc['2022-06':'2022-07'].Close.mean() #Mean of Price between time stamp of June to July 
```

```python
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
df.Close.resample('BM').mean().plot(kind='bar',color=['black','red','yellow','green','gray'],title='TATA STELL MARKET ANALYSIS',figsize=(10,5))
# Bar graph for Business Month end 
```

```python
df.Close.resample('M').mean().plot(kind='pie',figsize=(10,5)) #Pie Chart on a Montly Basis 
```

```python
 df.Close.resample('w').mean().plot(kind='line',color=['cyan'],figsize=(10,5))  # Line Graph on Weekly Basis 
```

```python
df.Close.plot(color=['Brown'] ,figsize=(10,5)) #On Daily Basis 
```
